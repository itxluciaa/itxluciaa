```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pickle

# Lista de géneros para iterar
generos = ['genero-Comedia', 'genero-Terror', 'genero-Acción', 'genero-Ciencia Ficción','genero-Romance',
           'genero-Thriller/Thriller Psicológico', 'genero-Drama', 'genero-Fantasía', 'genero-Otro', 'genero-Anime']

# Crear un diccionario para almacenar los modelos y sus resultados
modelos = {}

for genero in generos:
    # Definir la variable objetivo para el género actual
    y = EncuestaDummies[genero]
    X = EncuestaDummies[['Extrovertido', 'Empático', 'Reflexivo', 'Aventurero', 'Imaginativo',
       'Pasional', 'Adrenalítico', 'Idealista', 'emocion-Alegria',
       'emocion-Tensión', 'emocion-Reflexión', 'emocion-Sorpresa',
       'emocion-Empatía', 'emocion-Curiosidad', 'emocion-Tristeza',
       'emocion-Tranquilidad', 'emocion-Vergüenza ajena','Sexo_Hombre', 'Sexo_Mujer', 'Sexo_Otro', 'Edad_-18', 'Edad_+60',
       'Edad_18-30', 'Edad_30-40', 'Edad_40-60', 'Compañía_Acompañado',
       'Compañía_Solo', 'Reviews_Las leo pero no me suelen influir',
       'Reviews_Las leo y me influyen', 'Reviews_No las leo',
       'Plan_Ir a explorar un edificio abandonado', 'Plan_Ir a un scape room',
       'Plan_Ir a ver un atardecer',
       'Plan_Pasar el día en un parque de atracciones',
       'Plan_Pasar la tarde leyendo o viendo pelis']]
    
    # Dividir los datos en conjunto de entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)
    
    # Crear y entrenar el modelo
    arbol = DecisionTreeClassifier(random_state=1)
    arbol.fit(X_train, y_train)
    with open("{}-arbol.pkl".format(genero).replace('/', '-'), "wb") as f:
        pickle.dump(arbol, f)
    
    # Evaluar el modelo
    accuracy = accuracy_score(y_test, arbol.predict(X_test))
    
    # Guardar el modelo y su precisión en el diccionario
    modelos[genero] = {'modelo': arbol, 'accuracy': accuracy}

# Mostrar los resultados
for genero, data in modelos.items():
    print(f"Genero: {genero}, Accuracy: {data['accuracy']:.4f}")
```

```python
from sklearn.model_selection import cross_val_score

# Ajustes iniciales
scores = {}

for genero in generos:
    y = EncuestaDummies[genero]
    X = EncuestaDummies.drop(generos, axis=1)  # Asegúrate de excluir todas las columnas de género de las características

    arbol = DecisionTreeClassifier()
    # Aplicar validación cruzada
    cv_scores = cross_val_score(arbol, X, y, cv=6)  # Usar 5 pliegues por ejemplo
    scores[genero] = cv_scores.mean()  # Guardar la media de las puntuaciones de validación cruzada

# Mostrar los resultados de validación cruzada
for genero, score in scores.items():
    print(f"Genero: {genero}, CV Accuracy: {score:.4f}")
```
